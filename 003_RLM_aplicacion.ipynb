```python
df = pd.read_csv('./GE.csv')
df.head()
```

```python
df.shape
```

```python
df['Close(t)'].plot(figsize=(10, 7))
plt.title('Precio de GE' )
plt.ylabel('Precio')
plt.xlabel('Tiempo')
plt.grid(color='k')
plt.show()
```

```python
df = df.drop(columns=['Date', 'Date_col'])
df.head()
```

```python
N = df.shape[0]
N
```

```python
Y = df['Close(t)']
X = df.drop(columns=['Close(t)'])
```

```python
X_train = X.iloc[0: int(N*0.7)]
Y_train = Y.iloc[0: int(N*0.7)]

X_test = X.iloc[int(N*0.7):]
Y_test = Y.iloc[int(N*0.7):]
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
!pip install scikit-learn
```

```python
rlm = LinearRegression()
rlm.fit(X_train, Y_train)
```

```python
rlm.coef_
```

```python
rlm.intercept_
```

```python
Y_pred = rlm.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(Y_test, Y_pred)
```

```python
pd.DataFrame({'real': Y_test, 'pred':Y_pred}).plot(figsize=(10, 7))
```

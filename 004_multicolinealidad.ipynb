```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('./GE.csv')
df.head()
```

```python
df = df.drop(columns=['Date', 'Date_col'])
df.head()
```

```python
Y = df['Close(t)']
```

```python
variables = df.columns
variables
```

```python
d = {}

for variable in variables:
    if variable != 'Close(t)':
        Xj = df[variable] # definimos la variable indpendiente del modelo auxiliar en esta iteracion
        X = df.drop(columns=['Close(t)', variable])
        
        modelo = LinearRegression()
        modelo.fit(X, Xj) # calculamos los coeficientes del modelo auxiliar

        Xj_pred = modelo.predict(X) # calculamos las predicciones del modelo auxiliar
        Rj2 = r2_score(Xj, Xj_pred) # calculamos el coeficiente de correlacion del modelo auxiliar
        
        d[variable] = Rj2 # guardamos el coeficiente de correlacion
```
